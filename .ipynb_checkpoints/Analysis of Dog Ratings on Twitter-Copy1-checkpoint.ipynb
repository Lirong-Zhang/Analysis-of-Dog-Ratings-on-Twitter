```python
import tweepy
from tweepy import OAuthHandler
import json
from timeit import default_timer as timer
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import OrderedDict
import re
%matplotlib inline
```

```python
# Query Twitter API for each tweet in the Twitter archive and save JSON in a text file
# These are hidden to comply with Twitter's API terms and conditions
consumer_key = 'HIDDEN'
consumer_secret = 'HIDDEN'
access_token = 'HIDDEN'
access_secret = 'HIDDEN'

auth = OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)

api = tweepy.API(auth, wait_on_rate_limit=True)

# NOTE TO STUDENT WITH MOBILE VERIFICATION ISSUES:
# df_1 is a DataFrame with the twitter_archive_enhanced.csv file. You may have to
# change line 17 to match the name of your DataFrame with twitter_archive_enhanced.csv
# NOTE TO REVIEWER: this student had mobile verification issues so the following
# Twitter API code was sent to this student from a Udacity instructor
# Tweet IDs for which to gather additional data via Twitter's API
tweet_ids = df_1.tweet_id.values
len(tweet_ids)

# Query Twitter's API for JSON data for each tweet ID in the Twitter archive
count = 0
fails_dict = {}
start = timer()
# Save each tweet's returned JSON as a new line in a .txt file
with open('tweet_json.txt', 'w') as outfile:
    # This loop will likely take 20-30 minutes to run because of Twitter's rate limit
    for tweet_id in tweet_ids:
        count += 1
        print(str(count) + ": " + str(tweet_id))
        try:
            tweet = api.get_status(tweet_id, tweet_mode='extended')
            print("Success")
            json.dump(tweet._json, outfile)
            outfile.write('\n')
        except tweepy.TweepError as e:
            print("Fail")
            fails_dict[tweet_id] = e
            pass
end = timer()
print(end - start)
print(fails_dict)
```

```python
df0 = pd.read_csv('twitter-archive-enhanced.csv')
```

```python
image_pred = pd.read_csv('image-predictions.tsv', sep='\t')
```

Loop through the rows in tweet_json file and write the 'id', 'retweet_count', 'favorite_count' values into a dataframe.

```python
retweets = []
favorites = []
ids = []
with open('tweet-json.txt', 'r') as f:
    lines = f.readlines()
    for i in range(len(lines)):
        tweets_json = json.loads(lines[i])
        ids.append(tweets_json['id'])
        retweets.append(tweets_json['retweet_count'])
        favorites.append(tweets_json['favorite_count'])
tweets = OrderedDict({'tweet_id': ids, 'retweet_count': retweets, 'favorite_count': favorites})
df = pd.DataFrame.from_dict(tweets)
df
```

Merge the new dataframe with twitter-archive-enhanced.

```python
merged_df = df0.merge(df, on='tweet_id', how='left')
```

### Tidiness issues

1. Doggo, pupper, puppo, and floofer are all dog stages, should be in one column.
1. Some columns are irrelavant.
1. All tables can be combined as one.
2. rating_numerator and rating_denominator should be combined into one single column for the rating.

### Quality issues

1. In merged dataframe, tweet_id and timestamp are of wrong data types.
1. In image_pred dataframe, tweet_id is of wrong data type.
1. The 'retweet_count', 'favorite_count' values are missing in two rows in twitter-archive-enhanced.
1. Some name entries are clearly incorrect, like 'a', 'the'.
1. 'None' is regarded as a valid name.
1. Extracted the wrong numbers as the rating.
1. Some tweets have no rating.
1. Some dogs appear to be in multiple stages.
1. Some images are not dogs.

```python
# Make copies
merged_df_clean = merged_df.copy()
image_pred_clean = image_pred.copy()
```

**Define**

Doggo, pupper, puppo, and floofer are all dog stages, should be in one column.

**Code**

```python
merged_df_clean['stage'] = merged_df_clean[['doggo', 'floofer', 'pupper', 'puppo']].apply(lambda x: ''.join(x), axis=1)
merged_df_clean['stage'].replace("NoneNoneNoneNone","None ", inplace=True)
merged_df_clean['stage'].replace("doggoNoneNoneNone","doggo", inplace=True)
merged_df_clean['stage'].replace("NoneflooferNoneNone","floofer", inplace=True)
merged_df_clean['stage'].replace("NoneNonepupperNone","pupper", inplace=True)
merged_df_clean['stage'].replace("NoneNoneNonepuppo","puppo", inplace=True)
merged_df_clean['stage'].replace("doggoNonepupperNone","multiple", inplace=True)
merged_df_clean['stage'].replace("doggoflooferNoneNone","multiple", inplace=True)
merged_df_clean['stage'].replace("doggoNoneNonepuppo","multiple", inplace=True)
```

**Define**

Extracted the wrong numbers as the rating.

**Code**

```python
ratings_list = []
order_list = []
for i in range(merged_df_clean.shape[0]):
    if merged_df_clean.loc[i, 'rating_denominator'] != 10:
        ratings_list.append(re.findall(r'\d+\/\d+', merged_df_clean.loc[i, 'text']))
        order_list.append(i)
```

```python
ratings_list
```

```python
for n, i in enumerate(order_list):
    if n in [0, 4, 6, 8, 9, 18, 22]: # when len(ratings_list[n] == 2, the wrong rating is extracted)
        merged_df_clean.loc[i, 'rating_numerator'] = int(ratings_list[n][1][:-3])
        merged_df_clean.loc[i, 'rating_denominator'] = 10
```

### Explore the dog breeds that have been rated

There are more than 100 dog breeds that have been rated, the top 20 are shown in the pie chart, among which golden_retriever, Labrador_retriever and Pembroke are the top 3 most common ones.

### Explore the popularity over time

First we look at the retweet_count.

```python
plt.figure(figsize = (10, 8))
plt.plot(df_all.timestamp, df_all.retweet_count, marker = 'o', linestyle = '', ms = 2)
plt.yscale('log')
plt.title('Retweet count over time')
plt.xlabel('Time')
plt.ylabel('retweet_count (log)')
plt.show()
```

For favorite_count:

```python
plt.figure(figsize = (10, 8))
plt.plot(df_all.timestamp, df_all.favorite_count, marker = 'o', linestyle = '', ms = 2)
plt.yscale('log')
plt.title('Favorite count over time')
plt.xlabel('Time')
plt.ylabel('favorite_count (log)')
plt.show()
```

As can be seen above, the general popularity has been increasing constantly both in terms of retweet count and favorite count.

```python
df_all.sort_values('retweet_count', ascending=False)[:10].plot(x='p1', y='retweet_count', kind='barh');
```
